### Contenido del Dataset  
El dataset considerado contiene las siguientes columnas:  
* game: Nombre del videojuego.
* platform: Nombre de la versión de consola considerada. Muchos videojuegos son lanzados en distintas consolas por lo que habrá un registro por cada consola para cual haya sido lanzado ya que Metacritic divide a los juegos por versión de consola.
* developer: Nombre de la empresa desarrolladora del videojuego.
* genre: Género del videojuego.
* number_players: Número de jugadores que pueden participar en el videojuego.
* rating: Clasificación otorgada al videojuego por la ESRB.
* release_date: Mes, día y año en que fue lanzado el videojuego.
* postive_critics: Número de reseñas de críticos de videojuegos catalogadas como positivas por Metacritic.
* neutral_critics: Número de reseñas de críticos de videojuegos catalogadas como neutrales por Metacritic.
* negative_critics: Número de reseñas de críticos de videojuegos catalogadas como negativas por Metacritic.
* positive_users: Número de reseñas de usuarios catalogadas como positivas por Metacritic.
* neutral_users: Número de reseñas de usuarios catalogadas como neutrales por Metacritic.
* negative_users: Número de reseñas de usuarios catalogadas como negativas por Metacritic.
* metascore: Calificación promedio de la crítica. Metacritic convierte cada revisión en un porcentaje, ya sea matemáticamente a partir de la calificación dada, o lo que el sitio decide subjetivamente a partir de una revisión cualitativa. Antes de ser promediada, las puntuaciones se ponderan según la fama, el estatus y el volumen de las reseñas del crítico.
* user_score: Calificación promedio de los usuarios.

### Gráficos y análisis de resultados
**Inicio del código**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
path = "/home/nebelschwaden/Documents/2020A/PythonGR1/py-jaramillo-rojas-ronny-xavier/ProyectoPandas/metacritic_games.csv"
columnas = ['game','platform','developer','genre','number_players','rating','release_date',
            'positive_critics','neutral_critics','negative_critics','positive_users',
            'neutral_users','negative_users','metascore','user_score']
df = pd.read_csv(path,usecols = columnas)
```

**Número de videojuegos lanzados por plataforma**  
El siguiente gráfico denota la cantidad de videojuegos lanzados por plataforma entre los años 2011 y 2019, excluyendo a las consolas Xbox 360 y Playstation 3. Este gráfico nos permite conocer la preferencia de los publishers a la hora de lanzar un juego. Evidentemente, los publishers muestran una gran afinidad por las consolas Playstation 4 y Xbox One y por la PC. Si un juego es multiplataforma, el dataset contiene un registro de un juego por cada plataforma en la que se haya lanzado el juego por lo que esta comparación es válida.

```python
por_plataforma = df['platform'].value_counts()
pie_1 = plt.figure(figsize=(18,12))
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%\n({v:d})'.format(p=pct,v=val)
    return my_autopct
plt.pie(por_plataforma.values,labels=por_plataforma.index, autopct=make_autopct(por_plataforma.values))
plt.show()
```

**Top 10 desarrolladores con el mayor número de videojuegos publicados entre 2011 y 2019**  
El siguiente gráfico permite descubrir la frecuencia con la que diversos desarrolladores desarrollan un videojuego. Se puede observar que al desarrolladores 'Telltale Games' es aquella que mayor cantidad de videojuegos ha desarrollado de 2011 a 2019. Cabe recalcar que los videojuegos de esta empresa suelen contarse por episodios, es decir, por partes. Esto hace que se refleje una cantidad elevada de juegos publicados por su parte.

```python
unique_games = df[['game','developer']].drop_duplicates(subset='game', keep='first')
por_developer = unique_games['developer'].value_counts()[:10]
pie_2 = plt.figure(figsize=(10,7))
plt.pie(por_developer.values,labels=por_developer.index,autopct=make_autopct(por_developer.values))
plt.show()
```

**Top 10 géneros con mayor número de videojuegos**  
El siguiente gráfico permite visualizar los géneros más saturados en el mundo de los videojuegos. Se puede observar que el género más saturado es 'Acción' lo que deriva evidentemente en una mayor competencia si se piensa desarrollar un juego de este género.

```python
unique_by_genre = df[['game','genre']].drop_duplicates(subset='game', keep='first')
por_genre = unique_by_genre['genre'].value_counts()[:10]
bar = plt.figure()
ax = bar.add_axes([0,0,1,1])
ax.bar(por_genre.index,por_genre.values)
plt.xticks(rotation='vertical')
plt.show()
```

**Número de videojuegos por Rating (ESRB)**  
El siguiente gráfico muestra la cantidad de videojuegos por rating otorgado. Este rating es otorgado a cada videojuego por la ESRB e indica a qué público está dirigido dicho videojuego. De esta forma se puede ver que los desarrolladores tienen una afinidad equitativa por los públicos: jóven (Teen), adulto (Mature), general (Everyone), general mayor a diez años (Everyone +10). Se puede ver, evidentemente que el público menos 'deseado' es el de 'Sólo Adultos' (Adults Only).

**Número de lanzamientos por año**  
En el siguiente gráfico se puede observar el número de lanzamientos de videojuegos por año. En efecto, se denota que los lanzamientos por año han ido aumentando paulatinamente. Especialmente se observa los valores más altos entre los años 2015 y 2018. Este aumento se debe en parte al lanzamiento de las consolas de nueva generación en el año 2013; esto incentiva a los desarrolladores y publishers a ofrecer nuevos videojuegos para las nuevas generaciones de consolas.

```python
unique_by_date = df[['game','release_date']].drop_duplicates(subset='game', keep='first')
unique_by_date['release_date'] = unique_by_date['release_date'].str.split(",\s",2).str[1]
por_year = unique_by_date['release_date'].value_counts()
por_year = por_year.sort_index()
fig, ax = plt.subplots()
ax.plot(por_year.index, por_year.values) 
plt.show()
```

**Número de lanzamientos por mes entre los años 2011 y 2019**  
El siguiente gráfico permite observar la preferencia de los publishers para escoger la temporada de lanzamiento de un videojuego. Se denota que los publishers tienen afinidad por los meses de Marzo y Octubre para establecer sus fechas de lanzamientos de videojuegos.

```python
unique_by_month = df[['game','release_date']].drop_duplicates(subset='game', keep='first')
unique_by_month['release_date'] = unique_by_month['release_date'].str.split("\s",4).str[0]
por_month = unique_by_month['release_date'].value_counts()
cats = ['Jan', 'Feb', 'Mar', 'Apr','May','Jun', 'Jul', 'Aug','Sep', 'Oct', 'Nov', 'Dec']
por_month.index = pd.CategoricalIndex(por_month.index, categories=cats, ordered=True)
por_month = por_month.sort_index()
fig, ax = plt.subplots()
ax.plot(por_month.index, por_month.values) 
plt.show()
```

**Top 15 de videojuegos con mayor cantidad de reseñass positivas por parte de la crítica**  
El siguiente gráfico muestra los quince juegos con mayor acogida positiva por parte de la crítica. Cabe recalcar que estos no son los juegos más aclamados por la crítica ya que este gráfico se construye a partir del número de críticas catalogadas como positivas por parte del portal de Metacritic. Una crítica se considera como positiva si su calificación va desde 60 hasta 100. Como el dataset contiene múltiples registros de un mismo videojuego, se hizo una suma de todas las reseñas positivas por videojuego, es decir, se consideraron todas las reseñas de todos los videojuegos para todas las plataformas agrupandolos por videojuego. Se observa que el juego favorito de la crítica fue  DOOM (2016) ya que este posee el mayor número de críticas positivas considerando todas las plataformas en las que fué lanzado.

```python
all_ratings = df[['game','positive_critics']]
sum_ratings = all_ratings.groupby('game')['positive_critics'].sum()
sum_ratings = sum_ratings.sort_values(axis=0, ascending = False)[:15]
pie = plt.figure(figsize=(10,7))
plt.pie(sum_ratings.values,labels=sum_ratings.index,autopct=make_autopct(sum_ratings.values))
plt.show()
```

**Top 15 de videojuegos con mayor cantidad de reseñas neutras por parte de la crítica**  
El siguiente gráfico muestra los quince juegos con mayor acogida neutra por parte de la crítica. Al igual que el anterior gráfico, aquí se consideró la suma de todas las reseñas catalogadas como neutras por el portal de Metacritic. Una reseña es considera neutra si su valor está entre 40 y 60. Se observa que Anthem fue el videojuego con mayor acogida neutra por parte de la crítica. De manera general, estos videojuegos no fueron ni buenos ni malos según la crítica.

```python
all_neutral_ratings = df[['game','neutral_critics']]
sum_neutral = all_neutral_ratings.groupby('game')['neutral_critics'].sum()
sum_neutral = sum_neutral.sort_values(axis=0, ascending=False)[:15]
pie = plt.figure(figsize=(10,7))
plt.pie(sum_neutral.values,labels=sum_neutral.index,autopct=make_autopct(sum_neutral.values))
plt.show()
```

**Top 15 de videojuegos con mayor cantidad de reseñas negativas por parte de la crítica**  
El siguiente gráfico muestra los videojuegos con mayor cantidad de reseñas negativas por parte de la crítica. Una reseña es considerada negativa si su puntaje es menor a 30. Se observan diversos juegos que, de manera general, son considerados malos.
